```python
# Analysing costomer has table booking or not 
count = sns.countplot(x = 'Has Table booking',data = zomato, palette = 'Set2')
for label in count.containers:
    count.bar_label(label)
```

```python
# Analysing customer has online delivery or not
count = sns.countplot(x = 'Has Online delivery',data = zomato, palette = 'Set2')
for label in count.containers:
    count.bar_label(label)
```

```python
# Analysing the no of delivery in present
count = sns.countplot(x = 'Is delivering now',data = zomato, palette = 'Set2')
for label in count.containers:
    count.bar_label(label)
```

```python
pd.crosstab(zomato['Is delivering now'],zomato['Has Online delivery'])
```

```python
# Importing excel file
zomato_country= pd.read_excel('Country-Code.xlsx')
```

```python
# Merging Both
Final_zomato = pd.merge(zomato, zomato_country, on = 'Country Code', how = 'left')
```

```python
# Creating pivot table
Final_zomato.pivot_table(values = 'Aggregate rating', index = 'Rating text', columns = 'Country')
```

```python
sns.heatmap(Final_zomato.pivot_table(values='Aggregate rating', index = 'Rating text', columns = 'Country'), cmap='viridis')
```
